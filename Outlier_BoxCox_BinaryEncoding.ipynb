```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
outlier_df = pd.read_csv('/content/drive/MyDrive/EDA_finished.csv').drop(['Unnamed: 0'], axis = 1)
outlier_df.shape
```

```python
outlier_df.head()
```

```python
#method : scaling, log transformation, cube Root normalization, Box-Cox transformation
from scipy.stats import skew 
from scipy.stats import boxcox

outlier_df['PhysicalHealth'] = outlier_df['PhysicalHealth'] + 1
outlier_df['MentalHealth'] = outlier_df['MentalHealth'] + 1
outlier_columns = ['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']
for column in outlier_columns:
    print('이상치 적용하기 전 ' + column + ' 의 skewness : ', skew(outlier_df[column]))
    outlier_df[column] = boxcox(outlier_df[column])[0]
    print('이상치 적용한 후 ' + column + '의 skewness : ', skew(outlier_df[column]))
```

```python
import category_encoders as ce
import pandas as pd

ce_binary = ce.BinaryEncoder(cols = ['Race'], return_df = True)
ce_binary_encoded = ce_binary.fit_transform(outlier_df['Race'])
```

```python
outlier_df = pd.concat([outlier_df, ce_binary_encoded], axis = 1)
```

```python
outlier_df.to_csv('/content/drive/MyDrive/outlier_binaryencoding.csv')
```
