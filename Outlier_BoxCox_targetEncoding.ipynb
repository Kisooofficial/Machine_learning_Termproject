```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
outlier_df = pd.read_csv('/content/drive/MyDrive/EDA_finished.csv').drop(['Unnamed: 0'], axis = 1)
outlier_df.shape
```

```python
categories = outlier_df['Race'].unique()
targets = outlier_df['HeartDisease'].unique()
cat_list = []

for cat in categories:
    aux_dict = {}
    aux_dict['category'] = cat
    aux_df = outlier_df[outlier_df['Race'] == cat]
    counts = aux_df['HeartDisease'].value_counts()
    aux_dict['count'] = sum(counts)
    for t in targets:
        aux_dict['Race_' + str(t)] = counts[t]
    cat_list.append(aux_dict)

cat_list = pd.DataFrame(cat_list)
cat_list['genre_encoded_dumb'] = cat_list['Race_1'] / cat_list['count']

cat_list
```

```python
target_dict = {'White' : 0.098343, 'Black' : 0.075800, 'Asian' : 0.033279, 'American Indian/Alaskan Native' : 0.104391, 'Other' : 0.081352, 'Hispanic' : 0.053233}
outlier_df['Race_encoder'] = outlier_df['Race'].map(target_dict)
```

```python
#method : scaling, log transformation, cube Root normalization, Box-Cox transformation
from scipy.stats import skew 
from scipy.stats import boxcox

outlier_df['PhysicalHealth'] = outlier_df['PhysicalHealth'] + 1
outlier_df['MentalHealth'] = outlier_df['MentalHealth'] + 1
outlier_columns = ['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']
for column in outlier_columns:
    print('이상치 적용하기 전 ' + column + ' 의 skewness : ', skew(outlier_df[column]))
    outlier_df[column] = boxcox(outlier_df[column])[0]
    print('이상치 적용한 후 ' + column + '의 skewness : ', skew(outlier_df[column]))
```

```python
outlier_df.drop(['index'], axis = 1).to_csv('/content/drive/MyDrive/outlier_targetencoding.csv')
```
